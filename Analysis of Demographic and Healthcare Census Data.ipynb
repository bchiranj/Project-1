```python
#function 

def make_uniform(hospitals):
    
    #renaming states and UT's in uppercase
    
    census_df['State/UT'] = census_df['State/UT'].str.lower()
    
    c = census_df['State/UT'].replace(['laddakh'],['jammu and kashmir'])
    
    housing_df['State/UT'] = housing_df['State/UT'].str.lower()
    
    h = housing_df['State/UT'].replace(['laddakh'],['jammu and kashmir'])
    
    hospital_df['State/UT'] = hospital_df['State/UT'].str.lower().str.replace('*','')
    
    census_states = sorted(c.unique().tolist())
    
    housing_states = sorted(h.unique().tolist())
    
    hospital_states = sorted(hospitals['State/UT'].unique().tolist())
    
    #Checking for the number of states of in each of the datasets
    
    print(len(census_states),len(housing_states),len(hospital_states))
    
    states_with_diff_names=[]
    
    for i in range(1,len(census_states)):
        if census_states[i-1]!=housing_states[i-1] or census_states[i-1]!=hospital_states[i] or housing_states[i-1] != hospital_states[i]:
        
            c,hou,hos=census_states[i],housing_states[i],hospital_states[i]
            
            states_with_diff_names.append((c,hou,hos))
            
    print(states_with_diff_names) 
    
    hospital_df['State/UT'] = hospital_df['State/UT'].str.replace('&','and')
    
    hospital_df['State/UT'] = hospital_df['State/UT'].replace(['odisha','puducherry','delhi'],['orissa','pondicherry','nct of delhi'])
    
    return hospital_df['State/UT']
```

```python
def uniform_states(state):
    if state=='Nct of Delhi':
        state = 'Delhi'
    if state=='Puducherry':
        state =  'Pondicherry'
    if state=='Odisha':
        state='Orissa'
    if '&' in state:
        return state.replace("&", 'and')
    else:
        return state
```

```python
hospital_df['State/UT'] = make_uniform(hospital_df)
```
